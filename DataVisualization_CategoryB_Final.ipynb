# Data Visualization for Category B Datasets

```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading the dataset
df1 = pd.read_csv("TripB03.csv")
df2 = pd.read_csv("TripB08.csv")
df3 = pd.read_csv("TripB13.csv")
df4 = pd.read_csv("TripB22.csv")
df5 = pd.read_csv("TripB27.csv")
df6 = pd.read_csv("TripB30.csv")
df7 = pd.read_csv("TripB33.csv")
df8 = pd.read_csv("TripB37.csv")
```

```python
df1 = df1.fillna(method = 'bfill')
df2 = df2.fillna(method = 'bfill')
df3 = df3.fillna(method = 'bfill')
df4 = df4.fillna(method = 'bfill')
df5 = df5.fillna(method = 'bfill')
df6 = df6.fillna(method = 'bfill')
df7 = df7.fillna(method = 'bfill')
df8 = df8.fillna(method = 'bfill')
```

## SoC versus Time

```python
max_length = max(len(df) for df in [df1, df2, df3, df4, df5, df6, df7, df8])

time_data = []
soc_data = []

for i, df in enumerate([df1, df2, df3, df4, df5, df6, df7, df8]):
    soc = df['SoC [%]']
    time = df['Time [s]'] / 60

    time_data.append(time)
    soc_data.append(soc)
print(len(time_data))

plt.figure(figsize=(9,5))
plt.grid(True, linewidth=1, color='black', alpha=1)
lg = ['TripB03','TripB08','TripB13','TripB22','TripB27','TripB30','TripB33','TripB37']
colors = ['red','green','darkcyan','blue','brown','grey','orange','purple']

for i, (time, soc) in enumerate(zip(time_data, soc_data)):
    label = lg[i]
    plt.plot(time, soc, color=colors[i], linewidth=3, label=label)

plt.xlabel('Time [mins]', color='black', fontsize=14)
plt.ylabel('SoC [%]', color='black', fontsize=14)
plt.xlim((0,50))
plt.tick_params(axis='both', which='major', labelsize=14)

plt.legend(fontsize=14,ncol=2,loc="upper right",borderpad=0.35,handlelength=1.5,labelspacing=0.475,handletextpad=0.6)

plt.show()
```

```python
print(round(df1['SoC [%]'].describe(),4))
print(round(df2['SoC [%]'].describe(),4))
print(round(df3['SoC [%]'].describe(),4))
print(round(df4['SoC [%]'].describe(),4))
print(round(df5['SoC [%]'].describe(),4))
print(round(df6['SoC [%]'].describe(),4))
print(round(df7['SoC [%]'].describe(),4))
print(round(df8['SoC [%]'].describe(),4))
```

```python
print(round(df1['SoC [%]'].var(),4))
print(round(df2['SoC [%]'].var(),4))
print(round(df3['SoC [%]'].var(),4))
print(round(df4['SoC [%]'].var(),4))
print(round(df5['SoC [%]'].var(),4))
print(round(df6['SoC [%]'].var(),4))
print(round(df7['SoC [%]'].var(),4))
print(round(df8['SoC [%]'].var(),4))
```

## Current versus Time

```python
max_length = max(len(df) for df in [df1, df2, df3, df4, df5, df6, df7, df8]) 

time_data = []
current_data = []

for i, df in enumerate([df1, df2, df3, df4, df5, df6, df7, df8]):
    current = df['Battery Current [A]']
    time = df['Time [s]'] / 60

    time_data.append(time)
    current_data.append(current)

plt.figure(figsize=(9,5))
plt.grid(True, linewidth=1, color='black', alpha=1)
lg = ['TripB03','TripB08','TripB13','TripB22','TripB27','TripB30','TripB33','TripB37']
colors = ['red','green','darkcyan','blue','brown','grey','orange','purple']

for i, (time, current) in enumerate(zip(time_data, current_data)):
    label = lg[i]
    plt.plot(time, current, color=colors[i], linewidth=2, label=label)

plt.xlabel('Time [mins]', color='black', fontsize=14)
plt.ylabel('Current [A]', color='black', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=14)

plt.legend(fontsize=14,ncol=2,loc="lower right",borderpad=0.35,handlelength=2,labelspacing=0.475,handletextpad=0.6)

plt.show()
```
